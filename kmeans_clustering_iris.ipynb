# K-means clustering using the elbow method

```python
# importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# importing the iris dataset

dataset = pd.read_csv('Iris.csv')
X = dataset.iloc[:, 1:5].values
```

```python
dataset.head()
```

```python
#There are 3 types of species

dataset["Species"].unique()
```

```python
df1 = dataset[dataset["Species"] == "Iris-versicolor"]
```

```python
df2 = dataset[dataset["Species"] == "Iris-setosa"]
```

```python
df3 = dataset[dataset["Species"] == "Iris-virginica"]
```

```python
# Visualizing the dataset(Sepal Width vs Sepal length)

plt.scatter(df1["SepalLengthCm"], df1["SepalWidthCm"], s = 50, c = 'red', label='Iris-versicolor')
plt.scatter(df2["SepalLengthCm"], df2["SepalWidthCm"], s = 50, c = 'green', label='Iris-setosa')
plt.scatter(df3["SepalLengthCm"], df3["SepalWidthCm"], s = 50, c = 'blue', label='Iris-virginica')
plt.title('Sepal Width vs Sepal length')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
# Using the elbow method to find the optimal number of clusters

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters K')
plt.ylabel('Average Within-Cluster distance to Centroid (WCSS)')  
plt.show()
```

```python
# Applying k-means to the Iris dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualizing the clusters
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 50, c = 'red', label='Iris-versicolor')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 50, c = 'green', label='Iris-setosa')
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 50, c = 'blue', label='Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', 
            label = 'Centroids')
plt.title('Sepal Width vs Sepal length')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
# Visualizing the dataset(Petal Width vs Petal length)

plt.scatter(df1["PetalLengthCm"], df1["PetalWidthCm"], s = 50, c = 'red', label='Iris-versicolor')
plt.scatter(df2["PetalLengthCm"], df2["PetalWidthCm"], s = 50, c = 'green', label='Iris-setosa')
plt.scatter(df3["PetalLengthCm"], df3["PetalWidthCm"], s = 50, c = 'blue', label='Iris-virginica')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```

```python
# Visualizing the clusters
plt.scatter(X[y_kmeans == 2, 2], X[y_kmeans == 2, 3], s = 50, c = 'red', label='Iris-versicolor')
plt.scatter(X[y_kmeans == 1, 2], X[y_kmeans == 1, 3], s = 50, c = 'green', label='Iris-setosa')
plt.scatter(X[y_kmeans == 0, 2], X[y_kmeans == 0, 3], s = 50, c = 'blue', label='Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s = 100, c = 'yellow', 
            label = 'Centroids')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```
